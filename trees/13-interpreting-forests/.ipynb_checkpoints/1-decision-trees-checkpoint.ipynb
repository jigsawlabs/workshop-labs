# Decision Trees in Python

### Introduction

In the previous lesson, we worked with a sample dataset of customer leads and used a decision tree to predict whether or not a lead becomes a customer.  We did this by finding the attributes that most segment the groups into customers and non-customers.  And saw that if we first split our data into groups based on whether the our leads attended college, and categorize the remaining leads by whether they are under thirty we can construct a tree that predicts whether or nota lead will be a customer.  In this lesson, we'll see how we can perform this same procedure with Python.

### Formatting our Dataset

The first task, of course, is to load up our data.

```python
import pandas as pd
df = pd.read_csv('./customer_data.csv').iloc[:, 1:-1]
df
```

Now, because our machine learning models do not understand how to work with text, and can only process data in the form of numbers, our first step is to assign numbers to the categorical data we see above.  If we look at the datatypes of the dataframe above, we see the following:

```python
df.dtypes
```

So each of our features are currently of the datatype `object`, with each feature variable a string.  If we convert our features to type `category`, we can the change the feature variables to numbers.  Let's do it.

```python
X = df.iloc[:, :-1].astype('category')
```

```python
y = df.iloc[:, -1]
```

```python
X
```

Now that we have transformed the columns into categories, we can see each of the categories in the column. 

```python
X.attended_college.cat.categories
```

And if we like, we view a corresponding code for each entry in our categorical feature.

```python
X.attended_college.cat.codes
```

Now we can convert the categorical values in our dataframe with the following.

```python
X_coded = X.select_dtypes(include = 'category').apply(lambda x: x.cat.codes)
X_coded
```

Ok, so now we have all of our data converted into numbers, and we are ready to use sklearn's decision tree.

### Training a decision tree

We can train use a decision tree in sklearn by loading up the `DecisionTreeClassifier`.

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_coded, y)
```

Now the easiest way to understand the decision tree that has been trained, is to visualize it.  We can do so, with the `dtreeviz` library.

```python
viz = dtreeviz(dtc,
               X_coded.to_numpy(),
               y,
               target_name='is_customer',
               feature_names=X.columns,
               class_names=["customer", "non-customer"]
              )
              
viz   
```

```python
viz.view()
```

```python
X.attended_college.cat.categories
```

As we can see from the above visualization, the decision tree makes a split on if the lead attended college.  If the lead did attend college the tree immediate predicts the lead as a customer.  If the lead did not attend college, then the next test is if the lead is under thirty.  If not under thirty, it predicts a non-customer and if under thirty it predicts a customer.

We can also view how the decision tree works just working with the sklearn library.

```python
dtc.feature_importances_
```

The feature importances attribute shows us that the features used are the first two, `attended_college` and `under_thirty`.  We can pass through observations with variants of these features to see how the decision tree predicts them. 

```python
x_1 = [1, 1, 0,0]
x_2 = [1, 0, 0,0]

dtc.predict([x_1, x_2])
```
