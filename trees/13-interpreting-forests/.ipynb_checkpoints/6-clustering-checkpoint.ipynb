# Clustering Geographic Data 

### Introduction

### Loading the Data

```python
import pandas as pd
df_train = pd.read_feather('./bnb_train.feather')
df_X_train = df_train.drop(columns = ['price'])
y_train = df_train.price

df_val = pd.read_feather('./bnb_val.feather')
df_X_val = df_train.drop(columns = ['price'])
y_val = df_train.price

combined_df_X = pd.concat([df_X_train, df_X_val])
X = combined_df_X.loc[:, ['longitude', 'latitude']]
```

```python
df_X_train.shape
```

```python
from sklearn.cluster import KMeans

est = KMeans(n_clusters=9, init='random')
est.fit(X)
cluster_labels = est.labels_ 
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
plt.scatter(X.iloc[:, 1], X.iloc[:, 0], c=cluster_labels,  cmap=plt.cm.rainbow, s=35, linewidths=1, alpha=.3)
plt.xlabel('Long'), plt.ylabel('Lat')
plt.grid()
```

```python
combined_df_X.loc[:, 'cluster_number'] = cluster_labels
```

```python
X_train = combined_df_X.iloc[:11050]
```

```python
X_val = combined_df_X.iloc[11050:]
```

### Does it help?

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=40, max_features='log2')
rfr.fit(X_train, y_train)
rfr.score(X_val, y_val)
```

```python
from eli5.sklearn import PermutationImportance
perm = PermutationImportance(rfr).fit(X_val, y_val)
```

```python
import eli5
eli5.explain_weights_df(perm, feature_names = X_val.columns.to_list())[45:50]
```
