# Clean up Combine Data

```python
combine_df.info()
```

```python
drafted_cols = ['Team', 'Round', 'Pick']
combine_df[drafted_cols] = (combine_df['Drafted (tm/rnd/yr)']
                                      .str.split(' / ', expand=True)
                                      .iloc[:, :3])
```

```python
combine_df['Round'] = combine_df.Round.str.extract('(\d+)', 
                                                   expand=False)
combine_df['Pick'] = combine_df.Pick.str.extract('(\d+)', 
                                                 expand=False)
```

```python
# convert height to inches
ht = combine_df.Ht.str.split('-', expand=True).astype(int)
combine_df['Ht'] = ht[0]*12 + ht[1]
```

```python
combine_df.rename(columns={'40yd': 'Forty', '3Cone': 'Cone'},
                  inplace=True)
```

```python
combine_df.info()
```

```python
combine_df.head()
```

```python
combine_df.Pos.value_counts()
```

```python
combine_df.AV.fillna(0, inplace=True)
```

```python
# duplicate names will be appended with a 3 digit number
# so if John Smith shows up twice, then there
# should be two names, John Smith 001 and John Smith 002
# based my method from this answer
# https://stackoverflow.com/questions/43015345/pandas-dataframe-count-duplicate-rows-and-fill-in-column?rq=1
# https://stackoverflow.com/questions/37560548/how-to-change-value-of-second-duplicate-in-row
combine_df['occ_num'] = (combine_df.groupby('Player').cumcount() + 1).astype(str).str.zfill(2)
```

```python
dupl_mask = combine_df.Player.duplicated(keep=False)
```

```python
combine_df.iloc[[115,116, 849, 850]]
```

```python
combine_df.loc[dupl_mask,'Player'] = (combine_df.loc[dupl_mask, ['Player', 'occ_num']]
                                                .apply(lambda x: '-'.join(x), axis=1))
```

```python
combine_df.loc[dupl_mask,'Player'].head()
```

```python
cols_to_drop = ['School', 'College', 'Drafted (tm/rnd/yr)', 'occ_num']
combine_df.drop(cols_to_drop, axis=1, inplace=True)
```
