### Choosing the correct level of specificity

Let's work with our min_samples_leaf feature.  We can start with a minimum sample of two observations.  How do we know the correct number to use for max depth.  Well max depth is a hyperparameter, and so we can try various hyperparameters and score the performance against the test set.

```python
import numpy as np
num_sample_scores = []
for sample in range(1, 100):
    dtc = DecisionTreeRegressor(min_samples_leaf=sample)
    dtc.fit(X_train, y_train)
    score = dtc.score(X_test, y_test)
    num_sample_scores.append(np.array([sample, score]))
```

```python
num_sample_scores = np.array(num_sample_scores)
```

```python
from graph import trace_values, plot
```

```python
trace = trace_values(num_sample_scores[:, 0], num_sample_scores[:, 1])
plot([trace])
```

Now fitting our decision tree one more time, we arrive at the following.

```python
from sklearn.tree import DecisionTreeRegressor
dtc = DecisionTreeRegressor(min_samples_leaf=25)
dtc.fit(X_train, y_train)
dtc.score(X_test, y_test)
```

Let's view the tree.

```python
from dtreeviz.trees import *
viz = dtreeviz(dtc,
               X_train,
               y_train,
               target_name='disease progression',
               feature_names=dataset['feature_names'],
              )
              
viz   
```
